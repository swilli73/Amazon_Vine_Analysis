```python
import pandas as pd
```

```python
#Load in data
file_path = "vine_table.csv"
df = pd.read_csv(file_path)
df.head(10)
```

```python
#Filter df by total_votes greater than or equal to 20
new_df = df[df['total_votes']>=20]
new_df.head(10)
```

```python
#Filter the df to retrieve all the rows where the number of 
#helpful_votes divided by total_votes is equal to or greater than 50%.
new_df = new_df[df['helpful_votes']/df['total_votes']>=0.5]
new_df.tail(10)
```

```python
#Create new dataframe from new_df where vine=Y
yes_vine = new_df[new_df['vine']=='Y']
yes_vine.head()
```

```python
#Create new dataframe from new_df where vine=N
no_vine = new_df[new_df['vine']=='N']
no_vine.head()
```

```python
#fivestar = yes_vine[yes_vine['star_rating'] == 5.0]
#fivestar
yes_ids = yes_vine['review_id'].count()
yes_fives = (yes_vine['star_rating']==5.0).sum()
no_ids = no_vine['review_id'].count()
no_fives = (no_vine['star_rating']==5.0).sum()
```

```python
#Determine the total number of reviews, the number of 5-star reviews, and the percentage of 5-star reviews 
#for paid reviews
print(f"Total Paid Reviews: {yes_vine['review_id'].count()}")
print(f"Total 5-Star Paid Reviews: {(yes_vine['star_rating']==5.0).sum()}")
print(f"Percentage of 5-Star Paid Reviews: {yes_fives / yes_ids:.2%}")
```

```python
#Determine the total number of reviews, the number of 5-star reviews, and the percentage of 5-star reviews 
#for unpaid reviews
print(f"Total Unpaid Reviews: {no_vine['review_id'].count()}")
print(f"Total 5-Star Unpaid Reviews: {(no_vine['star_rating']==5.0).sum()}")
print(f"Percentage of 5-Star Unpaid Reviews: {no_fives / no_ids:.2%}")
```
